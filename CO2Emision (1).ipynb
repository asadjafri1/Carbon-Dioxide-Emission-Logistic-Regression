```python
import pandas
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df = pandas.read_csv("CO2Emissiondata.csv")
print(df.head(20))
X = df[['Weight', 'Volume']]
y = df['CO2']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[1500, 1140]])

print(predictedCO2)
```

```python
print(regr.coef_)
print(regr.intercept_)
```

```python
#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict(X)

plt.scatter(y,predictedCO2)
plt.scatter(y,y,color='r')
plt.show
```
